```python
mais_banidos = lol[["ban_1","ban_2","ban_3","ban_4","ban_5"]].stack().value_counts().reset_index().head(10)
```

```python
mais_banidos.columns = ["banidos","quantidade"]
mais_banidos
```

```python
import seaborn as  sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(16, 8))
plt.bar(mais_banidos["banidos"], mais_banidos["quantidade"], color = "#7b2cbf")
plt.xticks(rotation=90)
plt.xlabel("Campeões")
plt.ylabel("Vezes banido")
plt.xticks(rotation=0)
plt.title("Campeões mais banidos do competitivo League of Legends entre 2015 a 2018")
plt.show()
```

```python
ladoVermelho = lol.loc[lol['Team'] == "redBans"]
```

```python
mais_banidos_vermelho = ladoVermelho[["ban_1","ban_2","ban_3","ban_4","ban_5"]].stack().value_counts().reset_index().head(10)
mais_banidos_vermelho.columns = ["campeão","bans"]
mais_banidos_vermelho
```

```python
ladoAzul = lol.loc[lol['Team'] == "blueBans"]
ladoAzul
```

```python
mais_banidos_azul = ladoAzul[["ban_1","ban_2","ban_3","ban_4","ban_5"]].stack().value_counts().reset_index().head(10)
mais_banidos_azul.columns = ["campeão","bans"]
mais_banidos_azul
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))

axs[0].bar(mais_banidos_azul["campeão"], mais_banidos_azul["bans"])
axs[0].set_xlabel("Campeões")
axs[0].set_ylabel("Vezes banido")
axs[0].set_title("Campeões mais banidos do lado Azul do mapa")
axs[0].tick_params(axis='x', rotation=50)

axs[1].bar(mais_banidos_vermelho["campeão"], mais_banidos_vermelho["bans"], color= "red")
axs[1].set_xlabel("Campeões")
axs[1].set_ylabel("Vezes banido")
axs[1].set_title("Campeões mais banidos do lado Vermelho do mapa")
axs[1].tick_params(axis='x', rotation=50)

plt.show()
```
